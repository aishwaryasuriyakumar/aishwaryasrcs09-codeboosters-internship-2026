```python
import pandas as pd
df=pd.read_csv  ('student_performance.csv')
print("dataset loaded successfully")
```

```python
print("="*50)
print("      MINI PROJECT : STUDENT DATA EXPLORER")
print("="*50)


print('--------------------------------------------')
print('         [2] Dataset Objective')
print('--------------------------------------------')
print("Total students:", len(df))
print("Total columns:", len(df.columns))
print("Unique departments:",df['department'].nunique())

print('--------------------------------------------')
print('         [3] Department-wise Analysis ')
print('--------------------------------------------')
dept = df['department'].value_counts()
print(dept)

print('--------------------------------------------')
print('         [4] Score Analysis')
print('--------------------------------------------')
print("Highest Math:",df['math_score'].max())
print("Lowest Math:",df['math_score'].min())
print("Average Math:",df['math_score'].mean())
print("Average Programming:",df['programming_score'].mean())
print("Average Attendance:",df['attendance_percentage'].mean())

df['total_score'] = (df['math_score'] + df['science_score'] + df['english_score'] + df['programming_score'])

print('--------------------------------------------')
print('         [5] Gender wise Analysis')
print('--------------------------------------------')
gender = df.groupby('gender')['total_score'].mean()
print(gender)

print('--------------------------------------------')
print('         [6] Top performers')
print('--------------------------------------------')
top = df.sort_values(by='total_score',ascending=False)
print(top[['name','department','total_score']].head(5))

print('--------------------------------------------')
print('         [7] Attendance Analysis')
print('--------------------------------------------')
low = df[df['attendance_percentage'] < 75]
if len(low)==0:
    print("No students below 75%")
else:
    print(low[['name','department','attendance_percentage']])
```
